```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split 
import os
from sklearn.preprocessing import MinMaxScaler
```

```python
# Class work
dir_path = os.path.dirname(os.path.realpath('2022849446_21_11_22.ipynb'))
data = pd.read_csv(dir_path + '/Samsung_stock_price_dec_7.csv')
data
```

```python
df = data[['Date','Close']]
df
```

```python
data.set_index(pd.DatetimeIndex(data['Date']), inplace=True)
```

```python
df = df[['Close']]
df
```

```python
X = df.drop(['Date'],axis=1) 
Y = pd.to_datetime(df['Date'])

scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(X)

X_train, X_test, Y_train, Y_test = train_test_split(scaled_data, Y, test_size=.20, random_state=0)
 
def mape(Y_actual,Y_Predicted):
    mape = np.mean(np.abs((Y_actual - Y_Predicted)/Y_actual))*100
    return mape
```

```python
from sklearn.linear_model import LinearRegression
linear_model = LinearRegression().fit(X_train , Y_train)
 
lr_predict = linear_model.predict(X_test) 

lr_mape= mape(Y_test,lr_predict)

lr_mape
```
