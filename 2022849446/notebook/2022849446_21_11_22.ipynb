```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
# Class work
dir_path = os.path.dirname(os.path.realpath('2022849446_21_11_22.ipynb'))
data = pd.read_csv(dir_path + '/data.csv')
data
```

```python
fig, ax = plt.subplots(figsize=(16, 8))
x = data['UG'].values
y = data['Ph.D'].values
ax.plot(x,y)
plt.suptitle('Data', fontsize=18)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.20)
x_train_refit = x_train.reshape(-1,1)
x_test_refit = x_test.reshape(-1,1)
y_train_refit = y_train.reshape(-1,1)
y_test_refit = y_test.reshape(-1,1)

reg = LinearRegression().fit(x_train_refit, y_train_refit)
pred = reg.predict(x_test_refit)
pred
```

```python
reg.score(x_train_refit, y_train_refit)
```

```python
plt.plot(y_test, label='Actual')
plt.plot(pred, label='Prediction')
plt.xlabel('True value')
plt.ylabel('Pred value')
plt.title('Linear Regression')
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.20)
x_train_refit = x_train.reshape(-1,1)
x_test_refit = x_test.reshape(-1,1)
y_train_refit = y_train.reshape(-1,1)
y_test_refit = y_test.reshape(-1,1)

lgreg = LogisticRegression(random_state=0).fit(x_train_refit, y_train)
lgreg_pred = lgreg.predict(x_test_refit)
lgreg_pred
```

```python
lgreg.score(x_train_refit, y_train_refit)
```

```python
plt.plot(y_test, label='Actual')
plt.plot(lgreg_pred, label='Prediction')
plt.xlabel('True value')
plt.ylabel('Pred value')
plt.title('Logistic Regression')
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import Ridge
poly = PolynomialFeatures(3)

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.20)
x_train_refit = x_train.reshape(-1,1)
x_test_refit = x_test.reshape(-1,1)
y_train_refit = y_train.reshape(-1,1)
y_test_refit = y_test.reshape(-1,1)

x_train_poly = poly.fit_transform(x_train_refit)
x_test_poly = poly.fit_transform(x_test_refit)
y_train_poly = poly.fit_transform(y_train_refit)
y_test_poly = poly.fit_transform(y_test_refit)

ridge = Ridge(alpha=35)
ridge.fit(x_train_poly,y_train_refit)
ridge_pred = ridge.predict(x_test_poly)
ridge_pred
```

```python
plt.plot(y_test, label='Actual')
plt.plot(ridge_pred, label='Prediction')
plt.xlabel('True value')
plt.ylabel('Pred value')
plt.title('Polynomial Regression')
plt.show()
```
